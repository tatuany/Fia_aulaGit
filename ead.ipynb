## Covariancia e Correlação

A covariância é uma medida de como duas variáveis mudam uma em relação à outra.

Um número positivo significa que há um tendência de que, à medida que uma variável aumenta, a outra também aumenta.
Um número negativo significa que há um tendência de que, à medida que uma variável aumenta, a outra diminui.

A covariância entre duas colunas de um DataFrame pode ser calculada da seguinte maneira:

```python
df_casas.cov('preco', 'tam_terreno')
```

Por sua vez, a correlação é uma medida normalizada que é mais facil de entender, pois fornece medidas quantitativas da dependência estatística entre duas variáveis.

```python
df_casas.corr('preco', 'tam_terreno')
```

### Análise gráfica

Vamos analisar a relação entre as variáveis `ano_construcao` e `preco`

```python
df_s = df_casas.select('ano_construcao', 'preco')
df_s.display()
```

```python
import seaborn as sns

# O método toPandas() converte um DataFrame do Spark para um DataFrame do Pandas
sns.lmplot(y='preco', x='ano_construcao', data=df_s.toPandas())
```

** Qual o problema do comando anterior??**

O problema é que o `.toPandas()` irá recuperar todos os dados que estão distribuídos pelos nós do Cluster para o Driver.

Dependendo da quantidade de dados é bem provável que o programa irá travar. Lembre-se que a ideia de utilizar Spark é que a quantidade de dados que temos é maior do que a capacidade computacional.

O correto, antes de executar o `.toPandas()` é gerar uma amostra dos dados ou se certificar que o volume de dados que será transformado para o Pandas não seja muito grande.

```python
# gerando uma amostra
df_s = df_s.sample(withReplacement=False, fraction=0.5, seed=42)
sns.lmplot(y='preco', x='ano_construcao', data=df_s.toPandas())
```

O parâmetro `withReplacement` controla a exclusividade do resultado da amostra. Se tratarmos o conjunto de dados como um balde com bolinhas de gude, o valor `withReplacement=True` significa que ao retirar uma bolinha de gude de dentro do balde a mesma será retornada ao balde. Isso signficia que a bolinha de gude pode ser tirada novamente.

Assumindo que todos os valores sejam únicos no conjunto de dados:

- `withReplacement=True`, o mesmo elemento pode ser produzido mais de uma vez como resultado da amostra.
- `withReplacement=False`, cada elemento do conjunto de dados será amostrado apenas uma vez.

Podemos descatar os valores discrepantes (outliers) com 3 desvios padrão da média.

```python
valor_media = df_casas.select(F.mean('preco')).collect()[0][0]
valor_desvio_padrao = df_casas.select(F.stddev('preco')).collect()[0][0]

limite_superior = valor_media + (3 * valor_desvio_padrao)
limite_inferior = valor_media - (3 * valor_desvio_padrao)
```

## Tratando missing e removendo colunas

```python
# Remove linhas com qualquer valor NA - Abordagem Ingênua
df_casas.dropna().count() # note que ficou somente 7 linhas. Pois temos uma variável que tem muito missing.
```

```python
# Remover uma linha se tiver valores NA em 3 colunas
df_casas.dropna(thresh=3).count() # Não temos esse cenário, logo nenhuma linha foi removida.
```

```python
# Remover valores nulos olhando para apenas 1 coluna
df_casas.dropna(subset=['preco']).count()
```

```python
# Para remover colunas podemos
df_casas.drop('preco', 'ano_venda').display()
```

```python
# Contar a quantidade de missing
conta_nulos = [F.sum(F.when(F.isnull(col), 1).otherwise(0)).alias(f"qtd_nulos_{col}") for col in df_casas.columns]

df_casas.select(conta_nulos).display()
```

```python
# Vamos tratar o missing do `preco` substituindo pela média
media_preco = df_casas.select(F.mean('preco')).collect()[0][0]
media_preco
```

```python
df_casas = df_casas.fillna(media_preco, subset=['preco'])
```

```python
# Executando novamente a linha para verificar que os nulos na coluna preco foram substituídos pela média
df_casas.select(conta_nulos).select('qtd_nulos_preco').display()
```

Execute novamente a linha que verificar por missing. Note que a quantidade de missing da variável preço é 0.
